<h3 style='color:purple'>(1) Preprocessing: Detect face and eyes</h3>

#### When we look at any image, most of the time we identify a person using a face. An image might contain multiple faces, also the face can be obstructed and not clear. The first step in our pre-processing pipeline is to detect faces from an image. Once face is detected, we will detect eyes, if two eyes are detected then only we keep that image otherwise discard it.</h4>

#### Now how do you detect face and eyes?
We will use haar cascade from opencv

```python
# read an image from the test cases available
# the image has 3 dimensions ie y and y then rgb values
img = cv2.imread("./test_images/marissa_mayer.jpg")
img.shape
```

```python
# show the image
plt.imshow(img)
```

```python
# change the image to grey by removing the color
gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
gray.shape
```

```python
# the image is an array of numbers from 0-255
gray
```

```python
# plot the gray image
plt.imshow(gray,cmap="gray")
```

```python
# lets now detect the face and the eyes from the image
# refer to the Haar Cascades article about opencv-python for the code that does this
# refer to the haarcascades folder in the project, it has xml files for detecting these feature
face_cascade = cv2.CascadeClassifier("./opencv/haarcascades/haarcascade_frontalface_default.xml")
eye_cascade = cv2.CascadeClassifier("./opencv/haarcascades/haarcascade_eye.xml")
# gray is the gray image og marrisa mayer
faces = face_cascade.detectMultiScale(gray,1.3,5)
faces
```

The above will return an array of the face. It has 4 values ie 
x, y, width and height dimensions of the image (ie the face area)

```python
# lets now print out the face
# detect first face since faces is a 3D array 
(x,y,w,h) = faces[0]
x,y,w,h
```

```python
# draw the face out in a rectangular way in red color(rgb)
face_img = cv2.rectangle(img,(x,y),(x+w, y+h),(255,0,0),2)
plt.imshow(face_img)
```

```python
# Now draw the two eyes from the image
# code available in the opencv documentation
# we iterate through our faces (face_img) and then apply eye_cascade to get our region of interest
cv2.destroyAllWindows()
for (x,y,w,h) in faces:
    face_img = cv2.rectangle(img,(x,y),(x+w,y+h),(255,0,0),2)
    roi_gray = gray[y:y+h, x:x+w]
    roi_color = face_img[y:y+h, x:x+w]
    eyes = eye_cascade.detectMultiScale(roi_gray)
    for (ex,ey,ew,eh) in eyes:
        cv2.rectangle(roi_color,(ex,ey),(ex+ew,ey+eh),(0,255,0),2)
        

plt.figure()
plt.imshow(face_img, cmap='gray')
plt.show()
```

```python
# plot the cropped face from the image
plt.imshow(roi_color,cmap="gray")
```

So we will be cropping the face image for all the images and then storing the cropped faces per tech celebrity for model training

```python
# see the dimensions of the cropped image showing the region of interest, roi
cropped_img = np.array(roi_color)
cropped_img.shape
```

<h3 style='color:purple'> Preprocessing: Crop the facial region of the image</h3>

##### We will write a function to store all the cropped images in respective folders

```python
# we shall use image path to get the image
# we then read the image, convert it to gray, detect the face and 
# if we get number of eyes >=2, return the region of interest
def get_cropped_image_if_2_eyes(image_path):
    # print(f"Processing image: {image_path}")
    
    img = cv2.imread(image_path)
    
    # Check if the image is successfully loaded
    if img is None:
        # print(f"Error loading image: {image_path}")
        return None

    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    faces = face_cascade.detectMultiScale(gray, 1.3, 5)

    for (x, y, w, h) in faces:
        roi_gray = gray[y:y+h, x:x+w]
        roi_color = img[y:y+h, x:x+w]
        eyes = eye_cascade.detectMultiScale(roi_gray)
        if len(eyes) >= 2:
            return roi_color  # will return the first image

    # print(f"No two eyes detected in image: {image_path}")
    return None
```

```python
# this is the original image
original_image = cv2.imread("./test_images/marissa_mayer.jpg")
plt.imshow(original_image)
```

```python
# call the function on the image
cropped_image = get_cropped_image_if_2_eyes("./test_images/marissa_mayer.jpg")
plt.imshow(cropped_image)
```

```python
# first store the path of all the individual folders in a list
# then go thru all the subdirectories of the dataset folder and then store them in mg_dirs
import os
img_dirs = []
for entry in os.scandir(path_to_data):
    if entry.is_dir():
        img_dirs.append(entry.path)
```

Iterate through each of these images

```python
# create cropped_image_dirs to store cropped images for each of the celebrities
# split the image directory into 2 eg ./dataset/elon_musk' will be split into dataset and elon_musk
# then store the results in a list, [-1] will then return the last element in the list ie celeb name
cropped_image_dirs = []
celebrity_file_names_dict = {}
for img_dir in img_dirs:
    celebrity_name = img_dir.split("/")[-1]
    print(celebrity_name)
```

```python
# create cropped_image_dirs to store cropped images for each of the celebrities
# split the image directory into 2 eg ./dataset/elon_musk' will be split into dataset and elon_musk
# then store the results in a list, [-1] will then return the last element in the list ie celeb name
# iterate thru all the images of the resultant folders and then get a cropped image with both eyes
cropped_image_dirs = []
celebrity_file_names_dict = {}
for img_dir in img_dirs:
    # initialize count for images
    count = 1
    celebrity_name = img_dir.split('/')[-1]
    print(celebrity_name)

    celebrity_file_names_dict[celebrity_name] = []

    for entry in os.scandir(img_dir): #iterate and call function on each file and get cropped face if eyes are visible
        roi_color = get_cropped_image_if_2_eyes(entry.path)
        # if both eyes are visible as before, store image in cropped folder
        if roi_color is not None: #create a cropped folder containing path and celeb name
            cropped_folder = path_to_cr_data + celebrity_name
            # create the folder if not exist
            if not os.path.exists(cropped_folder):
                os.makedirs(cropped_folder)
                cropped_image_dirs.append(cropped_folder)
                # print the generated folder
                print("Generating cropped images in folder: ", cropped_folder)
            # generate the name of the file/image eg elon_musk1.png
            cropped_file_name = celebrity_name + str(count) + ".png"
            cropped_file_path = cropped_folder + "/" + cropped_file_name

            # save roi_color in the cropped file path
            cv2.imwrite(cropped_file_path, roi_color) # write path and img with roi
            # store the name of all image file paths into a dictionary
            celebrity_file_names_dict[celebrity_name].append(cropped_file_path)
            count+=1
            
```

<h2 style='color:purple'>(2) Feature Engineering</h2>
Here, we extract special features like eyes, nose, lips etc. Goal is to get X and y which we can use to train model

#### Do wavelet transformation
Helps extract important features from the image. Example below

```python
# function from stackoverflow
import numpy as np
import pywt
import cv2    

def w2d(img, mode='haar', level=1):
    imArray = img
    #Datatype conversions
    #convert to grayscale
    imArray = cv2.cvtColor( imArray,cv2.COLOR_RGB2GRAY )
    #convert to float
    imArray =  np.float32(imArray)   
    imArray /= 255;
    # compute coefficients 
    coeffs=pywt.wavedec2(imArray, mode, level=level)

    #Process Coefficients
    coeffs_H=list(coeffs)  
    coeffs_H[0] *= 0;  

    # reconstruction
    imArray_H=pywt.waverec2(coeffs_H, mode);
    imArray_H *= 255;
    imArray_H =  np.uint8(imArray_H)

    return imArray_H
```

```python
# lets try marissa mayer's cropped facial image
im_har = w2d(cropped_img,'db1',5)
plt.imshow(im_har, cmap='gray')
```

You can see above a wavelet transformed image (marissa_mayer) that gives clues on facial features such as eyes, nose, lips etc. This along with raw pixel image can be used as an input for our classifier. Notice that the area of each feature is differentiated </br>
To clearly understand this, research about frequency domain, time domain, signal processing, fourier transformation etc Details check Iman's youtube channel, 3Blue1Brown youtube channel.

#### Lets now do wavelet transform for all the images in the folders
We will vertically stack the images together ie both the raw and the transformed. This is serve a the input for the classifier

```python
# Lets go back and rebuild the dictionary coz we deleted some images
celebrity_file_names_dict = {}
for img_dir in cropped_image_dirs:
    celebrity_name = img_dir.split('/')[-1]
    file_list = []
    for entry in os.scandir(img_dir):
        file_list.append(entry.path)
    celebrity_file_names_dict[celebrity_name] = file_list
celebrity_file_names_dict
```

#### Images in cropped folder can be used for model training. We will use these raw images along with wavelet transformed images to train our classifier. Let's prepare X and y now

```python
# generate y as a number ie generate a class dictionary from the celebrit names
class_dict = {}
count = 0
for celebrity_name in celebrity_file_names_dict.keys():
    class_dict[celebrity_name] = count
    count = count+1
class_dict
```

```python
# define X and y
# X is the stacked image and y is the celebrity name
X = []
y = []
# we will iterate through the dictionary
# first loop iterates the image folders, 2nd loop iterates the images
for celebrity_name, training_files in celebrity_file_names_dict.items():
    for training_image in training_files:
        # read the image first using opencv
        img = cv2.imread(training_image)
        # if image does not exist ie for deleted images
        if img is None:
            continue
        # we need to do scaling to make sure images are of same size
        scalled_raw_image = cv2.resize(img,(32,32))
        # transform with w2d function
        img_har = w2d(img, "db1", 5)
        # scale the image
        scalled_img_har = cv2.resize(img_har,(32,32))
        # use numpy to vertically stack images ie 2D img with rgb values
        combined_img = np.vstack((scalled_raw_image.reshape(32*32*3,1),scalled_img_har.reshape(32*32,1)))
        # create X and y for model training
        X.append(combined_img)
        y.append(class_dict[celebrity_name])
```

```python
# total number of all images in all folders
len(X)
```

```python
# size of each image
len(X[0])
```

```python
# size calculated by as above
32*32*3 + 32*32
```

```python
# lets represent the images as float numbers
X = np.array(X).reshape(len(X),4096).astype(float)
X.shape
```

<h2 style='color:purple'>(3) Model Training</h2>
Now that we have our X and y ready, the next part is to do model training

#### We will use SVM with rbf kernel tuned  with heuristic finetuning
support vector machine is better with classfication

```python
# import relevant models
from sklearn.svm import SVC
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.pipeline import Pipeline
from sklearn.metrics import classification_report
```

```python
# split dataset into train and test sets
# then create a pipeline and scale the data
# randomly select parameters, we will later on do fine tuning
X_train, X_test, y_train, y_test = train_test_split(X,y,random_state=0)
pipe = Pipeline([("scaler",StandardScaler()),("svc",SVC(kernel="rbf",C=10))])
pipe.fit(X_train,y_train)
pipe.score(X_test,y_test)
```

68% accuracy is good. But better when training with neural network

```python
len(X_test)
```

Get the detailed report using the classification report

```python
print(classification_report(y_test,pipe.predict(X_test)))
```

<h2 style='color:purple'>(4) Parameter Fine Tuning</h2>

#### ``Use GridSearchCv to try out different models with different parameters.`` Goal is to come up with best model with best fine tuned parameters`

```python
from sklearn import svm
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import make_pipeline
from sklearn.model_selection import GridSearchCV
```

```python
# now define a dict of different candidate models to try.
model_params = {
    'svm': {
        'model': svm.SVC(gamma='auto',probability=True),
        'params' : {
            'svc__C': [1,10,100,1000],
            'svc__kernel': ['rbf','linear']
        }  
    },
    'random_forest': {
        'model': RandomForestClassifier(),
        'params' : {
            'randomforestclassifier__n_estimators': [1,5,10]
        }
    },
    'logistic_regression' : {
        'model': LogisticRegression(solver='liblinear',multi_class='auto'),
        'params': {
            'logisticregression__C': [1,5,10]
        }
    }
}
```

```python
# then run GridSearchCV on the dictionary
import pandas as pd
scores = []
best_estimators = {}
for algo, mp in model_params.items():
    pipe = make_pipeline(StandardScaler(),mp["model"])
    clf = GridSearchCV(pipe,mp["params"],cv=5,return_train_score=False)
    clf.fit(X_train,y_train)
    scores.append({
        "model": algo,
        "best_score": clf.best_score_,
        "best_params": clf.best_params_
    })
    best_estimators[algo] = clf.best_estimator_

df = pd.DataFrame(scores,columns=["model","best_score","best_params"])
df
```

```python
# return best estimators per model
best_estimators
```

```python
best_estimators["logistic_regression"].score(X_test,y_test)
```

```python
best_estimators["svm"].score(X_test,y_test)
```

```python
best_estimators["random_forest"].score(X_test,y_test)
```

Based on the above scores, we can see that logistic_regression has the best score on X_test and y_test still. Some data scientists use both GridSearchCv and (X_test and y_test) to choose which model is the best. Me I choose Logistic Regression, it does better for both tests

```python
# store best classifier
best_clf = best_estimators["logistic_regression"]
```

```python
# use a confusion matrix to data visualization
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test,best_clf.predict(X_test))
cm
```
